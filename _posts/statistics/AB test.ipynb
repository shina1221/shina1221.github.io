대응표본 t검정

!실험 이전의 집단과 실험 이후의 집단이 동일한 집단일 때 사용
!한 개인이 서로 다른 두 조건에서 짝을 지어 한 쌍이 연구대상이 되는 경우 사용

paired t test는 하나의 샘플에서 전과 후를 비교하는것
사람을 대상으로 한다면 A라는 같은 사람 1명
ex)특정 식이요법을 적용한 전과 후의 체중변화량 비교

t test는 두개의 샘플을 서로 비교하는 것으로 
사람을 대상으로 한다면 A와 B라는 사람 두명
ex)3교대를 하는 간호사 그룹 A와 외래 근무만 하는 간호사 그룹(B)간의 콜레스테롤 수치 비교

귀무가설은 평균값의 차이가 없다는것

```python
#참조
#https://data101.oopy.io/ab-test-python-or-without-code

from scipy import stats
design_A = [16.4, 12.6, 17.5, 18.8, 12.1, 13.2, 14.5, 12.4, 17.5, 25.4, 9.3, 10.4]
design_B = [12.1, 11.8, 14.7, 13.1, 13.8, 10.1, 9.1, 13.5, 11.2, 13.7]

#검정코드 실시
#ttest를 통해 두 집단의 평균이 같은지 파악

#A디자인과 B디자인이 있다고 할 때 평균 체류 시간은 A가 더 길었다
#여기서 A디자인의 체류시간이 더 긴것은 우연의 결과가 아닌가? 할때 ttest 사용

stats.ttest_ind(design_A, design_B, equal_var=False)
"""
A와 B의 분산이 너무 다르거나 수집된 데이터의 수가 너무 차이나면 계산방법이 달라야 함.
equal_var = False를 쓰면 welch의 t test 사용

최근 welch의 t test가 고전적인 student t test와 달리 데이터 크기(샘플 사이즈)와 분산성에 상관없이 안정된 p값을 도출한다는 연구가 제시됨
AB 테스트를 통해 수집한 데이터는 그 양이 지나치게 적지 않은한 정규성(극단적으로 크고 작은 값보다는 중간 정도의 값이 많은 성격) 정도는 보장되는 경우가 많은 것도 welch의 t test를 기본으로 사용하게 되는 이유 중 하나임.

equal_var는 두 집단의 variance가 같은지 다른지를 측정함. 
True일 때는 같다고, False일 때는 다르다고 보고 다른 테스트를 수행함.
"""

```

#독립성 검정(test of independence) : 서로 다른 요인들에 의해 분할되어 있는 경우 그 요인들이 관찰값에 영향을 주고 있는지 아닌지, 요인들이 서로 연관이 있는지 없는지를 검정. 두 개의 요인을 대상으로 함.

독립성 검정은 두개의 범주형 변수/요인이 서로 연관성이 있는지 상관이 있는지 독립적인지를 카이제곱 검정을 통해 통계적으로 판단하는 방법

ex)학력(초졸, 중졸, 고졸, 대졸, 대학원졸)이라는 범주형 변수와 연소득(하, 중, 상)이라는 범주형 변수간 서로 관련성이 있는지 관련이 없어 독립적인지 확인
  
//  
적합도 검정(goodness of fit test) : 관측값들이 어떤 이론적 분포를 따르고 있는지를 검정. 한개의 요인을 대상으로 함.

동질성 검정(test of homogeneity) : 관측값들이 정해진 범주 내에서 서로 비슷하게 나타나고 있는지를 검정. 속성 A,B를 가진 하위집단(subpopulation) 각각으로부터 정해진 표본의 크기만큼 자료를 추출하는 경우에 분할표에서 하위집단(subpopulation)의 비율이 동일한가를 검정. 두개의 요인을 대상으로 함.

#독립성 검정: 데이터셋이 독립성을 따르는지 검정
              귀무가설은 데이터셋이 정규분포를 따른다는 것

-카이제곱 검정
관찰된 빈도가 기대되는 빈도와 의미있게 다른지 여부를 검증하기 위해 사용.
독립성 검증:변인이 두 개 이상일 때 사용되며, 기대빈도는 두 변인이 서로 상관이 없고 독립적이라고 기대하는 것을 의미. 관찰빈도와의 차이를 통해 기대빈도의 진위여부를 밝힘.


```python
#독립성 검정
ser_1 = pd.DataFrame({'ID':['A','A','A','A','A','B','B','B','B','B'], 'YN':['Y','N','N','Y','N','Y','Y','N','Y','Y']})

#pandas의 pd.crosstab()으로 만든 결과물을 scipy의 chi2_contingency() 함수에 입력
from scipy.stats import chi2_contingency
df_crosstab=pd.crosstab(ser_1['ID'], ser_1['YN'])
chi2_contingency(df_crosstab)

#검정통계량, p-value, 자유도, 기대도수
```

#등분산성 검정 : 데이터셋의 분산이 같은지를 검정
                 귀무가설은 데이터셋의 분산이 같다

종류
-Bartlett's test
    from scipy.stats import bartlett
    비정규성에 민감하기 때문에 정규성을 따를 때에만 사용하는 검증 방법

-Levene's test
    from scipy.stats import levene
    정규분포와 관계없이 연속적인 분포로부터 데이터를 구했을 때 사용하는 검정 방법

정규성을 검정할 수 없을 때에는 Levene 검정을 사용하고 정규성이 확실하면 Bartlett 검정 사용

```python
import pandas as pd

#등분산성 검정
ser_1 = pd.Series([2,5,3,4,6,2,3])
ser_2 = pd.Series([7,3,6,5,2,6,7])

#정규성 검정
from scipy.stats import shapiro
print('정규성: ',shapiro(ser_1),shapiro(ser_2))
#둘다 정규성을 따르지 않음 따라서 이 떄 lenene 검정법을 사용해야 함.
#다만 둘 중 하나만 정규성을 따른다면 어느검정법을 따라야 하는지 모르곘음

#Bartlett's test
from scipy.stats import bartlett#(바틀렛)
print('bartlett:', bartlett(ser_1, ser_2))

#Levene's test
from scipy.stats import levene#(레빈)
print('levene:', bartlett(ser_1, ser_2))
#등분산성을 만족하지 못함
```

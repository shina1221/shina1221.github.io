훈련샘플: 데이터셋을 나타내는 테이블의 행(=관측, 레코드, 인스턴스, 예시)  
훈련: (=모델 피팅) 모수 모델의 경우 파라미터 추정과 비슷  
특성: 데이터 테이블이나 데이터 행렬의 열 (= 예측변수, 변수, 입력, 속성, 공변량)  
타깃: (=결과, 출력, 반응변수, 종속변수, (클래스)레이블, 정답)  
손실 함수: (=비용 함수), 일부 자료에서는 손실 함수를 하나의 데이터 포인트에 대해 측정한 손실로 사용하고, 비용함수는 전체 데이터셋에 대해 계산한 손실(평균 또는 합)로 사용

많은 머신러닝 알고리즘에서 최적의 성능을 내려면 선택된 특성이 같은 스케일을 가져야 함.
특성을 [0,1]범위로 변환하거나 평균이 0이고 단위 분산을 가진 표준정규분포로 변환하는 경우가 많음.

>>차원축소  
일부 선택된 특성은 상관관계가 높아 어느정도 중복 정보를 가질 수 있으므로 이때는 차원 축소기법을 활용해 특성을 저차원 부분 공간으로 압축.
특성 공간의 차원을 축소하면 저장 공간이 덜 필요하고 학습 알고리즘을 더 빨리 실행할 수 있음.
데이터에 관련없는 특성(또는 잡음)이 매우 많을 경우 즉, 신호 대 잡음비가(Signal to Noise// SNR)가 낮은 경우 차원 축소를 이용해 모델의 예측 성능을 높이기도 함.

머신러닝 알고리즘이 훈련 데이터셋에서 잘 작동하고 새로운 데이터 셋에서도 잘 일반화되는지 확인하기 위해서 데이터 셋을 훈련셋과 테스트셋으로 나눠 훈련셋으로 머신러닝 모델을 훈련해 최적화 한 뒤, 최종 모델을 평가할 때 테스트셋 사용.

##특성 스케일 조정과 차원 축소같은 단계에서 사용한 파라미터는 훈련 데이터셋만 사용해 얻은 것임.
  나중에 동일한 파라미터를 테스트 데이터셋은 물론 새로운 모든 데이터 샘플을 변환하는 데 사용.
  그렇지 않으면 테스트 셋에서 측정한 성능은 과도하게 낙관적인 결과를 보임.
  개인해석)훈련셋 범위를 기준으로 테스트 셋 변환

>> 모델검증
현실에서는 가장 좋은 모델을 훈련하고 선택하기 위해 최소한 몇 가지 알고리즘을 비교해야함.
여러 모델을 비교하기 전에 먼저 성능을 측정할 지표를 결정해야함.
분류에서 주로 사용되는 지표는 정확도(accuracy)임. 정확도는 정확히 분류된 샘플 비율임.

훈련 데이터셋에서 최적의 모델을 선택한 후에는 테스트 데이터셋을 사용해 이전에 본 적 없는 데이터에서 얼마나 성능을 내는지 예측해 일반화 오차 예상.
이 성능에 만족한다면 해당 모델을 사용해 데이터 예측에 활용.

>> 교차검증  
모델 선택에 테스트 데이터 셋을 사용하지 않고 최종 모델을 평가하려고 따로 보관한다면?
데스트 데이터 셋과 실제 데이터에서 어떤 모델이 잘 동작할지 어떻게 알 수 있을까?
교차 검증에서는 모델의 일반화 성능을 예측하기 위해 훈련 데이터를 훈련 데이터셋과 검증 데이터셋으로 더 나눔.

>> 하이퍼파라미터
머신 러닝 라이브러리들에서 제공하는 알고리즘의 기본 하이퍼파라미터는 최적이라 할 수 없음.
따라서 모델의 성능을 상세하게 조정하기 위해 하이퍼파라미터를 조정함.
